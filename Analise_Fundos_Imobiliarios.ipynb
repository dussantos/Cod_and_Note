```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
# web scraping / Obtendo dados
FII_url = requests.get("https://www.fundsexplorer.com.br/ranking")
```

```python
FII_bs4 = BeautifulSoup(FII_url.text, "html.parser")
```

```python
FII_table = FII_bs4.findAll(attrs = {'id':'scroll-wrapper'})
```

```python
table = FII_table[0].findAll('table')
```

```python
df = pd.read_html(str(FII_table[0]))[0]
```

```python
df['P/VPA'] = df['P/VPA']/100
```

```python
# Dados de todos os FIIs disponíveis
df.columns
```

```python
indicadores=['Códigodo fundo', 
             'Setor', 
             'Preço Atual',
             'P/VPA',
             'DY (12M)Acumulado',
             'VacânciaFísica',
             'VacânciaFinanceira',
             'Liquidez Diária',
             'QuantidadeAtivos',
             'PatrimônioLíq.',
             'VPA',
             'Dividendo',
             'DividendYield',
           
]
```

```python
FII_df = df[indicadores]
```

```python
str(list(FII_df['Códigodo fundo'].sort_values()))
```

```python
# Organização dos dados estraídos
# Faremos cópia do dataframe para manipulá-lo com mais liberdade
#FII_df_backup = FII_df.copy()
```

```python
# Eliminando os R$  
FII_df = FII_df.replace(to_replace= r"^R\$ ", value = "", regex = True)
```

```python
# Eliminando os "%" 
FII_df = FII_df.replace(to_replace= r"%$", value = "", regex = True)
```

```python
# Removendo os pontos dos separadores decimais
FII_df['PatrimônioLíq.'] = FII_df['PatrimônioLíq.'].str.replace(".","")
FII_df['Preço Atual'] = FII_df['Preço Atual'].str.replace(".","")
FII_df['Dividendo'] = FII_df['Dividendo'].str.replace(".","")
FII_df['VPA'] = FII_df['VPA'].str.replace(".","")
```

```python
# Substituir as "," por "." e Transformar as variasveis em float
FII_df['PatrimônioLíq.'] = FII_df['PatrimônioLíq.'].str.replace(",",".").astype(float)
FII_df['Preço Atual'] = FII_df['Preço Atual'].str.replace(",",".").astype(float)
FII_df['Dividendo'] = FII_df['Dividendo'].str.replace(",",".").astype(float)
FII_df['VPA'] = FII_df['VPA'].str.replace(",",".").astype(float)
FII_df['DividendYield'] = FII_df['DividendYield'].str.replace(",",".").astype(float)
FII_df['DY (12M)Acumulado'] = FII_df['DY (12M)Acumulado'].str.replace(',','.').astype(float)
FII_df['VacânciaFísica'] = FII_df['VacânciaFísica'].str.replace(',','.').astype(float)
FII_df['VacânciaFinanceira'] = FII_df['VacânciaFinanceira'].str.replace(',','.').astype(float)
```

```python
FII_df.dropna(subset=['Preço Atual','DividendYield','Setor'], inplace = True)
FII_df
```

```python
# Fundos vendidos abaixo do valor de mercado
FII_df['Fundos_VPA'] = FII_df["P/VPA"] < 1
FII_VPA = FII_df[FII_df['Fundos_VPA']==True]
FII_VPA.sort_values(by='Códigodo fundo')
```

```python
# Dividend Yield maior que 10%
FII_VPA_Select= FII_VPA[FII_VPA['DY (12M)Acumulado']>10]
FII_VPA_Select.sort_values(by='Códigodo fundo')
```

```python
# média "mean" e desvio padrão "std"
pivot_table_FIIs = FII_df.groupby("Setor").agg(["mean","std"])
pivot_table_FIIs
```

```python
# Função para possíveis oportunidades de mercado
def oportunidade_setor (df, setor="Outros", label_setor="Setor"):

  pivot_table_FIIs = FII_df.groupby('Setor').agg(['mean','std'])

  df_setor = df[df[label_setor].isin([setor])]

  filter_ = \
            (df_setor['QuantidadeAtivos'] > 5) &\
            (df_setor['Liquidez Diária'] > 800) &\
            (df_setor['P/VPA'] < 1) &\
            (df_setor['DY (12M)Acumulado'] > pivot_table_FIIs.loc[setor, ('DY (12M)Acumulado','mean')]) 
  print('média do setor Yield: {}'.format(pivot_table_FIIs.loc[setor, ('DY (12M)Acumulado','mean')]))
  print('média do setor p/VPA: {}'.format(pivot_table_FIIs.loc[setor, ('P/VPA','mean')]))
  print('média do setor Ativos: {}'.format(pivot_table_FIIs.loc[setor, ('QuantidadeAtivos','mean')]))
    
  return df_setor[filter_]
```

```python
oportunidade = oportunidade_setor(FII_df, setor='Outros')
oportunidade.sort_values('DY (12M)Acumulado', ascending=False, inplace=True)
oportunidade
```
